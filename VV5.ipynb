```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

# Q1

```python
image1 = cv2.imread('/content/1.png', cv2.COLOR_BGR2RGB)
image2 = cv2.imread('/content/2.png', cv2.COLOR_BGR2RGB)
image3 = cv2.imread('/content/3.png', cv2.COLOR_BGR2RGB)
image4 = cv2.imread('/content/4.png', cv2.COLOR_BGR2RGB)
image5 = cv2.imread('/content/5.png', cv2.COLOR_BGR2RGB)
image6 = cv2.imread('/content/6.jpg', cv2.COLOR_BGR2RGB)
```

```python
def Kmeans(img, K):
  pixel = np.float32(img.reshape((-1,3)))
  criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0) # criteria to stop iterations. one is when accuracy (1.0) is reached, other when (10) itereations are done
  ret,label,center=cv2.kmeans(pixel, K, None, criteria, 10, cv2.KMEANS_PP_CENTERS)
  center = np.uint8(center)
  result = center[label.flatten()]
  result = result.reshape((img.shape))
  return result
```

```python
r1 = Kmeans(image1, 3)
r2 = Kmeans(image1, 6)
r3 = Kmeans(image1, 8)

plt.figure(figsize = (15,15))

plt.subplot(141), plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(142), plt.imshow(cv2.cvtColor(r1, cv2.COLOR_BGR2RGB)), plt.title('K=3'), plt.axis('off')
plt.subplot(143), plt.imshow(cv2.cvtColor(r2, cv2.COLOR_BGR2RGB)), plt.title('K=6'), plt.axis('off')
plt.subplot(144), plt.imshow(cv2.cvtColor(r3, cv2.COLOR_BGR2RGB)), plt.title('K=8'), plt.axis('off')
plt.show()
```

# Q2



```python
image = cv2.imread('/content/boat.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
otsu_threshold, otsu_result = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
normal_threshold, normal_result = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.figure(figsize=(15,15))
plt.subplot(131), plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB)), plt.title('Original Image'), plt.axis('off')
plt.subplot(132), plt.imshow(otsu_result, cmap = 'gray'), plt.title('Otsu Thresholding'), plt.axis('off')
plt.subplot(133), plt.imshow(normal_result, cmap='gray'), plt.title('Normal Thresholding'), plt.axis('off')
plt.show()
```

```python
print(otsu_threshold)
print(normal_threshold)
```

**Conclusion**

*In normal thresholding, we set the threshold value to 127 as we dont know what could be more optimal whereas in otsu thresholding, it automatically determines & sets the suitable threshold which in this case is 132. Otsu thresholding is giving better results than normal thresholding.*
