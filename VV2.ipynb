```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('/content/pexels-simon-berger-1323550.jpg')
```

```python
print(image)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
L = image.max()
negative = L - image
plt.imshow(cv2.cvtColor(negative, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
image2 = cv2.resize(image, (256,256))
image = cv2.copyMakeBorder(image2, 2, 2, 2, 2, cv2.BORDER_CONSTANT, None, value = 0)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
img2 = image.copy()

# Traverse image portions alone
for i in range(2, image.shape[0] - 2):
    for j in range(2, image.shape[1] - 2):
        # for a 5x5 window sliding
        neighbors = []

        # Fetch all neighbours for the middle element of a 5x5 matrix
        for k in range(i - 2, i + 3):
            for l in range(j - 2, j + 3):
                neighbors.append(image[k, l])

        minimum = np.amin(neighbors, axis=0)
        img2[i, j] = minimum

# Display the original and processed images using matplotlib

plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Minimum Neighbor Operation')
plt.axis('off')

plt.show()
```
