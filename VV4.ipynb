# Q2

```python
image2 = cv2.imread('/content/original.jpg')
```

```python
colored2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
gray2 = cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)

gaussian2 = cv2.GaussianBlur(colored2, (5, 5), 0)
median2 = cv2.medianBlur(colored2, 5)
laplacian2 = cv2.Laplacian(gray2, cv2.CV_64F)
bilateral2 = cv2.bilateralFilter(colored2, 9, 75, 75)
box2 = cv2.boxFilter(colored2, -1, (5, 5))

plt.figure(figsize=(15, 10))
plt.subplot(231), plt.imshow(image2), plt.title('Original Image'), plt.axis('off')
plt.subplot(232), plt.imshow(gaussian2), plt.title('Gaussian Filter'), plt.axis('off')
plt.subplot(233), plt.imshow(median2), plt.title('Median Filter'), plt.axis('off')
plt.subplot(234), plt.imshow(laplacian2, cmap='gray'), plt.title('Laplacian Filter'), plt.axis('off')
plt.subplot(235), plt.imshow(bilateral2), plt.title('Bilateral Filter'), plt.axis('off')
plt.subplot(236), plt.imshow(box2), plt.title('Box Filter'), plt.axis('off')
plt.show()
```
